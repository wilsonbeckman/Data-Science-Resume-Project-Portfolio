```python
#Wanted to look at the time difference between plays
#This is difference from minutes played because it includes the time I wasn't listening to music
def calculate_time_diff(df, col_name):
    df['time_diff_minutes'] = None
    
    for i in range(1, len(df)):
        time_diff = (df.loc[i, col_name] - df.loc[i-1, col_name]).total_seconds() / 60
        df.at[i, 'time_diff_minutes'] = time_diff
    
    df['time_diff_minutes'] = df['time_diff_minutes'].astype(float)
    
    return df

history = calculate_time_diff(history, 'ts')

history.head()
```

```python
#time of day column is meant to assign meaning to the time of day I'm listening to music
from datetime import datetime, timedelta

early_morning_start = datetime.strptime('04:00:01', '%H:%M:%S').time()
early_morning_end = datetime.strptime('07:30:00', '%H:%M:%S').time()
morning_start = datetime.strptime('07:30:01', '%H:%M:%S').time()
morning_end = datetime.strptime('12:00:00', '%H:%M:%S').time()
afternoon_start = datetime.strptime('12:00:01', '%H:%M:%S').time()
afternoon_end = datetime.strptime('15:30:00', '%H:%M:%S').time()
evening_start = datetime.strptime('15:30:01', '%H:%M:%S').time()
evening_end = datetime.strptime('20:00:00', '%H:%M:%S').time()
night_start = datetime.strptime('20:00:01', '%H:%M:%S').time()
night_end = datetime.strptime('23:59:59', '%H:%M:%S').time()
late_night_start = datetime.strptime('00:00:00', '%H:%M:%S').time()
late_night_end = datetime.strptime('04:00:00', '%H:%M:%S').time()

def calculate_period_of_day(df):
    df['Period_of_Day'] = None
    for i, row in df.iterrows():
        if row['Time'] >= early_morning_start and row['Time'] <= early_morning_end:
            df.at[i, 'Period_of_Day'] = 'Early Morning'
        elif row['Time'] >= morning_start and row['Time'] <= morning_end:
            df.at[i, 'Period_of_Day'] = 'Morning'
        elif row['Time'] >= afternoon_start and row['Time'] <= afternoon_end:
            df.at[i, 'Period_of_Day'] = 'Afternoon'
        elif row['Time'] >= evening_start and row['Time'] <= evening_end:
            df.at[i, 'Period_of_Day'] = 'Evening'
        elif row['Time'] >= night_start and row['Time'] <= night_end:
            df.at[i, 'Period_of_Day'] = 'Night'
        else:
            df.at[i, 'Period_of_Day'] = 'Late Night'
    
    return df
```

```python
history = calculate_period_of_day(history)
```

It's a lot easier to read these time series charts
